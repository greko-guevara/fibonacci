```python
from math import cos, log, e, sin, radians
import matplotlib.pyplot as plt
import matplotlib
import pandas as pd
```

```python

#input 
n=10

x=0 #contador 
a=1 #segundo valor Fibonacci
f=[1] # primer valor en list 

while x < (n-1):
    f.append(a) # escritura de cada uno de los calores 
    a=f[x]+f[x+1] # función fibonacci
    x+=1 # contador 

print (f)
```

```python

#golden ratio 
phi=(1+5**0.5)/2
bb=log(phi)/90 #Constante para gráfico spiral
aa=0.001 #Constante para gráfico spiral
gold_angle=360-(360/phi)
print (phi)
print (bb)
```

```python

# grafico espiral 
df=pd.DataFrame()
nn= range(0,1000)
for j in nn:
    df.at[j,'Delta']=j
    df.at[j,'cos']=cos(radians(j))
    df.at[j,'sin']=sin(radians(j))


df["x"]=aa*df['cos']*e**(bb*df['Delta'])
df["y"]=aa*df['sin']*e**(bb*df['Delta'])

fig, ax = plt.subplots( figsize=(5, 5))
ax.plot(df["x"],df["y"])
ax.grid(True)
ax.set_title('model of Fibonacci´s spiral ')
plt.show()

```

```python

# grafico de flor de girasol 

df["dist_center"]=df["Delta"]**0.5
theta=0
for j in nn:
    df.at[j,"theta"]=theta
    theta=theta+gold_angle
    df.at[j,'ccos']=cos(radians(theta))
    df.at[j,'ssin']=sin(radians(theta))

df["xx"]=df["dist_center"]*df['ccos']
df["yy"]=df["dist_center"]*df['ssin']

fig, ax = plt.subplots( figsize=(15, 15))
ax.plot(df["xx"],df["yy"])
ax.grid(True)
ax.set_title('model of sunflower seed arrangement')
plt.show()
print(df)

```
